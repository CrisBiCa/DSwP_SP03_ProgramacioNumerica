## Flights Delay DataSet description:

This dataset is composed by the following variables:
From :https://www.kaggle.com/code/adveros/flight-delay-eda-exploratory-data-analysis

- Year 2008
- Month 1-12
- DayofMonth 1-31
- DayOfWeek 1 (Monday) - 7 (Sunday)
- DepTime actual departure time (local, hhmm)
- CRSDepTime scheduled departure time (local, hhmm)
- ArrTime actual arrival time (local, hhmm)
- CRSArrTime scheduled arrival time (local, hhmm)
- UniqueCarrier unique carrier code
- FlightNum flight number
- TailNum plane tail number: aircraft registration, unique aircraft identifier
- ActualElapsedTime in minutes
- CRSElapsedTime in minutes
- AirTime in minutes
- ArrDelay arrival delay, in minutes: A flight is counted as "on time" if it operated less than 15 minutes later the scheduled time shown in the carriers' Computerized Reservations Systems (CRS).
- DepDelay departure delay, in minutes
- Origin origin IATA airport code
- Dest destination IATA airport code
- Distance in miles
- TaxiIn taxi in time, in minutes
- TaxiOut taxi out time in minutes
- Cancelled *was the flight cancelled
- CancellationCode reason for cancellation (A = carrier, B = weather, C = NAS, D = security)
- Diverted 1 = yes, 0 = no
- CarrierDelay in minutes: Carrier delay is within the control of the air carrier. Examples of occurrences that may determine carrier delay are: aircraft cleaning, aircraft damage, awaiting the arrival of connecting passengers or crew, baggage, bird strike, cargo loading, catering, computer, outage-carrier equipment, crew legality (pilot or attendant rest), damage by hazardous goods, engineering inspection, fueling, handling disabled passengers, late crew, lavatory servicing, maintenance, oversales, potable water servicing, removal of unruly passenger, slow boarding or seating, stowing carry-on baggage, weight and balance delays.
- WeatherDelay in minutes: Weather delay is caused by extreme or hazardous weather conditions that are forecasted or manifest themselves on point of departure, enroute, or on point of arrival.
- NASDelay in minutes: Delay that is within the control of the National Airspace System (NAS) may include: non-extreme weather conditions, airport operations, heavy traffic volume, air traffic control, etc.
- SecurityDelay in minutes: Security delay is caused by evacuation of a terminal or concourse, re-boarding of aircraft because of security breach, inoperative screening equipment and/or long lines in excess of 29 minutes at screening areas.
- LateAircraftDelay in minutes: Arrival delay at an airport due to the late arrival of the same aircraft at a previous airport. The ripple effect of an earlier delay at downstream airports is referred to as delay propagation.

**Was the flight delayed?**

In order to do a new column registering if the flight was or wasnot delayed, I will compute all the different Types of Delays ('CarrierDelay', 'WeatherDelay', 'NASDelay', 'SecurityDelay', 'LateAircraftDelay') see if any of this happends and put the new column with True if yes False if not.

For now, I will not use ArrDelay and DeptDelay because I assume its already in the columns of the reason of the delay, but when I have the column with all the delays computed I'll check if this assumtion is true

```python
flights_df.columns
```

```python
flights_df.loc[flights_df["CarrierDelay"] != 0]
```

```python
##I create the condition that I want to look for, so the delay columns different from 0 since each of this column is in miunutes of delay
##Then I create the new column with True if any of the delayd columns is diferent from 0 and fals if all of them are 0
condition = (flights_df['CarrierDelay'] != 0) | (flights_df['WeatherDelay'] != 0) | (flights_df['NASDelay'] != 0) | (flights_df['SecurityDelay'] != 0) | (flights_df['LateAircraftDelay'] != 0) 
flights_df['FlightDelayed'] = np.where(condition, True, False)
```

```python
flights_df
```

### A table with the Carriers with more delay

To check this I want to group the data by carrier and sort by the amout of delays

```python
##I groupby Unique Carrier, sum the values so I can have a sum of the True's in Flight Delayed,
##and sort descending by Flight delayed
flights_df.groupby('UniqueCarrier').sum().sort_values(['FlightDelayed'], ascending=False)
```

```python
##If I just one the key column of the group by and the Flight delay I also can do it like this
CarrierDelayed = flights_df.groupby('UniqueCarrier')['FlightDelayed'].sum()
CarrierDelayed.sort_values(ascending=False)
```

We can See that the Carrier that accumulates more delays is WN and the one with less is AQ

```python
##We can directly use describe() methode to perform a set of aggregations that describe each group in the data
##I creat a new df to be able to sort it
CarrierStats = flights_df.groupby('UniqueCarrier')['FlightDelayed'].describe()
CarrierStats.sort_values('freq', ascending= False)
```

Here we see the basics statistic of the discribe for the table. The count referes to the number of times the Carrier had true or False, uniques is that we just have 2 objects (true and False) te top referes to the one of the to that appears more times and the frequency is the number of times it appears.

#### Which is the longest route?

```python
##Here I groupby two columns Origine and Destination, because I want to check which are the longest routs
##I do it in 2 steps, to see what is doing, first grouping by and aggregating by sum
flights_df.groupby(['Origin', 'Dest']).sum()
```

```python
##Second step is to sort by the distance column and I put the table in a new df
RouteDistance = flights_df.groupby(['Origin', 'Dest']).sum().sort_values('Distance', ascending=False)
##Here I just show the 3 key columns and the distance on the new df, and just the first 10 rows
RouteDistance[['Distance']][:10]
```

In the previous table we see the routes with the longest distance

#### Which is the route with more delay?

```python
##Here I groupby two columns Origine and Destination, and sort by the flight delay column and I put the table in a new df
RouteDistance = flights_df.groupby(['Origin', 'Dest']).sum().sort_values('FlightDelayed', ascending=False)
##Here I just show the 2 key columns and the Flight delayed on the new df, and just the first 10 rows
RouteDistance[['FlightDelayed']][:10]
```
